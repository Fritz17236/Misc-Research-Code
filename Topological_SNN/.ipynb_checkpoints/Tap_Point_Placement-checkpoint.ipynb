```python

def cluster_to_grid(clusters, somas, n):
    '''
    Given soma-encoder clusters, return n x n matrix grid where grid[i,j] = c
    if soma[i,j] is assigned to cluster c, also returns the centroids of each cluster on the grid
    '''
    cl_vals = np.zeros((n,n))
    
    centroid_locs = {}
    
    for c in clusters:  
        sum_row = 0
        sum_col = 0
        clust_size = 0
        es = clusters[c][0]
        
        for e in es: 
            soma = somas[tuple(e)]
            row,col = idx_to_grid(soma, n, order='F')
            sum_row += row
            sum_col += col 
            clust_size += 1
            cl_vals[row,col] = c
        
        centroid_locs[c] = (sum_row // clust_size, sum_col // clust_size)    
            
    return cl_vals, centroid_locs

def draw_cut(row,col,direction,n, scale=1):
    '''
    given a location (row, col) of pixels on a grid, and a
    direction (up='u', down='d', left='l' right='r'), draw a 
    black edge at pixel (row,col) on the [direction] side of the 
    pixel. Scale determines the width of a single edge relative to the 
    pixels they are written over.
    row col referes to the upper left corner of the pixel
    '''
    
    # if dir up or down line will be horizontal
    # if dir left or right line will be vertical
    
    assert(direction in ['u', 'd', 'l', 'r']), 'direction must be one of u, d, l, or r'
    cuts = []
    if direction=='u':
        cuts.append(plt.axhline(y=row-.5*scale, xmin=col/n, xmax=(col+scale)/n, color='k'))
        
    if direction=='d':
        cuts.append(plt.axhline(y=row+.5*scale, xmin=col/n, xmax=(col+scale)/n, color='k'))
        
    if direction=='l':
        cuts.append(plt.axvline(x=col-.5*scale, ymin=1-row/n, ymax=1-(row+scale)/n, color='k'))
        
    if direction=='r':
        cuts.append(plt.axvline(x=col+.5*scale, ymin=1-row/n, ymax=1-(row+scale)/n, color='k'))
    
def get_diffuser_cuts_by_cluster(grid):
    '''
    given a grid with assigned clusters, return a list of all the cuts between clusters.
    this is a list of vertices (row, col, dir) between where two different clusters border
    each other 
    '''
    cuts = []
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            
            this_cluster = grid[i,j]
            if i > 0:
                if grid[i-1,j] != this_cluster:
                    cuts.append((i,j,'u'))
                
            if i < grid.shape[0]-1:
                if grid[i+1,j] != this_cluster:
                    cuts.append((i,j,'d'))
            
            if j > 0:
                if grid[i,j-1] != this_cluster:
                    cuts.append((i,j, 'l'))
            if j < grid.shape[1]-1:
                if grid[i,j+1] != this_cluster:
                    cuts.append((i, j, 'r'))
            
    return cuts

def get_diffuser_cuts_by_angle(grid, somas, encs, n, thresh = 45):
    '''
    Return a list of diffuser cuts where a cut is made between two adjacent 
    somas if their encoding vectors have an angle (in degrees) greater than 
    thresh.  
    '''
    cuts = []
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            
            
            this_cluster = grid[i,j]
            this_soma = grid_to_idx((i,j), n, order='F')
            this_enc = encs[this_soma]
            
            if i > 0:
                up_enc = encs[grid_to_idx((i-1,j), n, order='F')]
                if angle(this_enc, up_enc, radians=False) > thresh:
                    cuts.append((i,j,'u'))
                
            if i < grid.shape[0]-1:
                down_enc = encs[grid_to_idx((i+1,j), n, order='F')]
                if angle(this_enc, down_enc, radians=False) > thresh:
                    cuts.append((i,j,'d'))
            
            if j > 0:
                left_enc = encs[grid_to_idx((i,j-1), n, order='F')]
                if angle(this_enc, left_enc, radians=False) > thresh:
                    cuts.append((i,j, 'l'))
                    
            if j < grid.shape[1]-1:
                right_enc = encs[grid_to_idx((i,j+1), n, order='F')]
                if angle(this_enc, right_enc, radians=False) > thresh:
                    cuts.append((i, j, 'r'))
            
    return cuts

plt.figure()
clusters = cluster_binary(encs)
grid, center_locs = cluster_to_grid(clusters,somas,n)
cuts = get_diffuser_cuts_by_angle(grid, somas, encs, n, thresh=45)
for cluster in center_locs.keys():
    row,col = center_locs[cluster]
    grid[row,col] = -1
plt.imshow(grid)


for (row, col, direction) in cuts:
    draw_cut(row, col, direction, n)

    
tap_matrix = np.zeros((N, d))
for c in clusters.keys():
    row,col = center_locs[c]
    c_centroid = clusters[c][1]
    center_soma = grid_to_idx((row, col), n, order='F')
    tap_matrix[center_soma, :] = c_centroid 
    
    

plt.figure()
plt.imshow(tap_matrix[:,0].reshape((n,n),order='F'))

plt.figure()
plt.imshow(tap_matrix[:,1].reshape((n,n),order='F'))

plt.show()
# 2) diffusive breaks are inserted between clusters
    # because diffuser cuts can only be inserted between 4x4 tiles of neurons, we have 2 options:
    #  a) treat each 4x4 tile of neurons as a single neuron and run our algorithm on 1/4 the given neurons 
    #  OR
    #  b) split the grid into 4x4 tiles and interpolate each tile based on whether 2 or more tiles from a given cluster are
    # present 
    
    # for now we go with a. our assignments is for a 64 x 64 neuron grid, so we instead use 256x256 neuron pool with each 4x4 tile mapped to a soma from our algo

# Both x and y should be relative to the top left corner of the pool. Possible
# directions are “right”, “left”, “up”, and “down”. The edges of the pool are cut automatically, to
# prevent currents from diffusing across the pool edges. The diffuser can only be cut at edges
# of a 4x4 tile of neurons.
```
